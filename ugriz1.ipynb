```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
hdu = fits.open('filter_curves.fits')
def printugriz1(ii):
    hdu_x=hdu[ii].data
    flux=hdu_x.field(1)
    wavelength=hdu_x.field(0)
    xi=wavelength
    yi=flux
    plt.plot(xi,yi,'',color = 'black',linewidth=0.5)
    plt.xlabel('wavelength(0.1nm)')
    plt.ylabel('flux(relative)')
for i in range(1,6):
    printugriz1(i)
```
